```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def estimate_distribution(k:int, predictor, generator) -> float:
    sample_size:int=1000
    theta:float = 10
    n:int = 1000

    result:float=0
    for i in range(0, n):
        sample = generator(theta,sample_size)
        prediction = predictor(sample, k)
        result += pow(theta - prediction, 2)
    result /= n
    result = pow(result, 0.5)
    return result

def print_graph(predictor, generator):
    k_ammount:int = 100
    X=range(1, k_ammount)
    Y=[estimate_distribution(x, predictor, generator) for x in X]
    plt.plot(X, Y)
    plt.show()
```

```python
# uniform distribution
def generate_uniform_distribution(theta:int, n:int) -> np.array:
    return np.random.uniform(0, theta, n)

def predict_uniform_parameter(sample: np.array, k:float) -> float:
    n:int=len(sample)
    sum:float=0
    for i in range(0, n):
        sum += pow(sample[i],k)
    sum /= n
    return pow((k + 1)*sum, 1/k)

print_graph(predict_uniform_parameter, generate_uniform_distribution)
```

```python
# exponential distribution

def generate_exponential_distribution(theta:int, n:int) -> np.array:
    return np.random.exponential(theta, n)

def predict_exponential_parameter(sample: np.array, k:float) -> float:
    n:int=len(sample)
    sum:float=0
    for i in range(0, n):
        sum += sample[i]**k
    sum /= n
    return pow(sum / math.factorial(k), 1/k)
    
print_graph(predict_exponential_parameter, generate_exponential_distribution)
```
